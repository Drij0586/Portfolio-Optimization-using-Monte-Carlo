```python
from pandas_datareader import data as pdr
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
import numpy as np
```

```python
Stocks = ['AAPL' ,'MSFT' ,'AMZN' ,'IBM']
```

```python
end = dt.datetime.now()
start = dt.datetime(2023 , 1 , 1)
```

```python
import yfinance as yfin
yfin.pdr_override()
df = pdr.get_data_yahoo(Stocks , start , end)
```

```python
df.head()
```

```python
data = df['Adj Close']
```

```python
data.head()
```

```python

weights = np.random.random(4)
weights = weights / weights.sum()
weights
```

```python
percentage_change = data.pct_change()
```

```python
percentage_change
```

```python
annual_return = percentage_change.mean()*252
expected_return = np.sum(annual_return*weights) 
```

```python
expected_return
```

```python
exp_vol = np.sqrt(np.dot(weights.T , np.dot(np.cov(annual_return), weights)))
```

```python
exp_vol
```

```python
sharpe_ratio = expected_return / exp_vol
sharpe_ratio
```

```python
N = 10000
weights = np.zeros((N , 4))
exp_return = np.zeros(N)
exp_vol = np.zeros(N)
sharpe_ratios = np.zeros(N)

for i in range(10000):
    weight = np.random.random(4)
    weight = weight / np.sum(weight)
    weights[i] = weight
    
    exp_return[i] = np.sum(annual_return*weight)
    exp_vol[i] = np.sqrt(np.dot(weight.T , np.dot(np.cov(annual_return), weight)))
    sharpe_ratios[i] = exp_return[i] / exp_vol[i]
```

```python
sharpe_ratios.max()
sharpe_ratios.argmax()
```

```python
weights[sharpe_ratios.argmax()]
```

```python
plt.figure(figsize=(14,7))
figure , axis  = plt.subplots()
axis.scatter(exp_vol , exp_return , c = sharpe_ratios)
axis.set_xlabel('Expected Volatility')
axis.set_ylabel('Expected Return')
```
